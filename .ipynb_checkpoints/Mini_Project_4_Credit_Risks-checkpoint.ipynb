```python
#Be a bit more clever and check for optimal number of clusters
scores = []
neighbors = range(1,20)

for n in neighbors:
    knn = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    results = knn.score(X_test,y_test)
    scores.append(results)
    
print(scores)
plt.plot(neighbors,scores)
```

```python
#Redefine KNN with number of clusters showing highest score on test set
knn = KNeighborsClassifier(n_neighbors=13).fit(X_train,y_train)
```

```python
#Check model result with validation set
result = knn.score(X_validate,y_validate)
print("Accuracy = ",result)
```

```python
#Print confusion matrix using validation set
predictions = knn.predict(X_validate)
print("Confusion Matrix: \n",confusion_matrix(y_validate,predictions))
```
