```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('datasets/insurance.csv')
df.head()
```

```python
df.shape
```

```python
df.dropna()
```

```python
df['sex'] = LabelEncoder().fit_transform(df['sex'])
df['smoker'] = LabelEncoder().fit_transform(df['smoker'])
df['region'] = LabelEncoder().fit_transform(df['region'])
df.head()
```

```python
df.describe()
```

```python
df["smoker"].value_counts()[1]
```

```python
df["charges"] = StandardScaler().fit_transform(df["charges"].values.reshape(-1,1))
df["bmi"] = StandardScaler().fit_transform(df["bmi"].values.reshape(-1,1))
df["age"] = StandardScaler().fit_transform(df["age"].values.reshape(-1,1))
df.head()
```

```python
sns.pairplot(df.iloc[:,0:-1])
```

```python
corr = df.iloc[:,0:-1].corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr,cmap="coolwarm",annot=True)
```

```python
plt.figure(figsize=(20,5))
ax=sns.boxplot(data=df.iloc[:,0:-1])
```

```python
#Define variables for classifcation
y = df["charges"].values
X = df.iloc[:,0:-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
X_validate, X_test, y_validate, y_test = train_test_split(X_test,y_test,test_size=0.5,random_state=0)
```

```python
num_trees = range(1,100)
scores_forest = []
results = 0
best_score_forest = 0

for nt in num_trees:
    
    model_forest = RandomForestRegressor(criterion='mse',max_leaf_nodes=100,n_estimators=nt)
    model_forest.fit(X_train,y_train.ravel())
    results = model_forest.score(X_test,y_test)
    scores_forest.append(round(results,2))
    
    if results > best_score_forest:
        best_score_forest = results
        best_nt = nt
        best_model_forest = model_forest
        
print("R2:",scores_forest)
print("Optimal number of trees:",best_nt)       
```

```python
plt.plot(num_trees,scores_forest)
plt.xlabel('Number of trees')
plt.ylabel('R2 Score')
plt.title('Random Forest Training')
```

```python
y_pred = best_model_forest.predict(X_validate)
print("MSE:" + str(mean_squared_error(y_validate,y_pred)))
print("MAE:" + str(mean_absolute_error(y_validate,y_pred)))
print("R2:" + str(r2_score(y_validate,y_pred)))
```

```python
kernels = ['linear', 'rbf', 'poly']
regularization = [1, 10, 100, 1000]
scores_svr = np.zeros([4,4])
results = 0
best_score_svr = 0

for k in range(len(kernels)):
    for c in range(len(regularization)):
        
        model_svr = SVR(kernel=kernels[k],C=regularization[c],gamma='scale',epsilon=0.1)
        model_svr.fit(X_train, y_train.ravel())
        results = model_svr.score(X_test,y_test)
        scores_svr[k,c] = (round(results,2))
        
        if results > best_score_svr:
            best_score_svr = results
            best_kernel = kernels[k]
            best_regularization = regularization[c]
            best_model_svr = model_svr


print("R2:",scores_svr.reshape(1,-1))
print("Optimal kernel:",best_kernel)
print("Optimal regularization:",best_regularization)
```

```python
for k in np.arange(len(kernels)):
    plt.plot(regularization,scores_svr[k,:],label=kernels[k])
plt.title('SVR Training')
plt.legend(loc='upper left')
plt.xlabel('regularization')
plt.ylabel('R2 score')
```

```python
y_pred = best_model_svr.predict(X_validate)
print("MSE:" + str(mean_squared_error(y_validate,y_pred)))
print("MAE:" + str(mean_absolute_error(y_validate,y_pred)))
print("R2:" + str(r2_score(y_validate,y_pred)))
```

```python
max_depth = range(1,100)
scores_tree = []
results = 0
best_score_tree = 0

for md in max_depth:

    model_tree = DecisionTreeRegressor(criterion='mae',max_depth=md,random_state=0)
    model_tree.fit(X_train, y_train)
    results = model_tree.score(X_test,y_test)
    scores_tree.append(round(results,2))
    
    if results > best_score_tree:
        best_score_tree = results
        best_md = md
        best_model_tree = model_tree
        
print("R2:",scores_tree)
print("Optimal tree depth:",best_md)
```

```python
plt.plot(max_depth,scores_tree)
plt.xlabel('Max depth of tree')
plt.ylabel('R2 Score')
plt.title('Decision Tree Training')
```

```python
y_pred = best_model_tree.predict(X_validate)
print("MSE:" + str(mean_squared_error(y_validate,y_pred)))
print("MAE:" + str(mean_absolute_error(y_validate,y_pred)))
print("R2:" + str(r2_score(y_validate,y_pred)))
```
