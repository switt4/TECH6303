```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans, MeanShift
```

```python
df_wisconsin = pd.read_csv('datasets/wisconsin_breast_cancer_data.csv')
df_wisconsin.head()
```

```python
df_wisconsin.drop(['id','Unnamed: 32'],axis=1,inplace=True)
df_wisconsin['diagnosis'] = LabelEncoder().fit_transform(df_wisconsin['diagnosis'])
df_wisconsin.iloc[:,1:-1] = StandardScaler().fit_transform(df_wisconsin.iloc[:,1:-1])
df_wisconsin.head()
```

```python
df_wisconsin.to_csv('datasets/wisconsin_breast_cancer_data_refined.csv')
```

```python
df_insurance = pd.read_csv('datasets/insurance.csv')
df_insurance.head()
```

```python
df_insurance['sex'] = LabelEncoder().fit_transform(df_insurance['sex'])
df_insurance['smoker'] = LabelEncoder().fit_transform(df_insurance['smoker'])
df_insurance['region'] = pd.get_dummies(df_insurance['region'])
df_insurance['age'] = StandardScaler().fit_transform(df_insurance['age'].values.reshape(-1,1))
df_insurance['bmi'] = StandardScaler().fit_transform(df_insurance['bmi'].values.reshape(-1,1))
df_insurance.head()
```

```python
features_wisconsin = df_wisconsin.iloc[:,1:-1]
```

```python
clusters = range(1,31)
inertia = []
results = 0

for c in clusters:
    kmeans = KMeans(n_clusters=c, tol = 0.01, random_state=0).fit(features_wisconsin.values)
    results = kmeans.inertia_
    inertia.append(round(results,2))

plt.plot(clusters,inertia, marker='*',c='blue')
plt.xlabel('Number of clusters')
plt.ylabel('inertia')

#print("The inertia for each cluster:",inertia)
```

```python
kmeans_optimal = KMeans(n_clusters=5, tol = 0.01, random_state=0).fit_predict(features_wisconsin.values)
features_wisconsin["kmeans"] = kmeans_optimal.tolist()
features_wisconsin.head()
```

```python
plt.scatter(features_wisconsin["radius_mean"],features_wisconsin["texture_mean"],c=features_wisconsin.kmeans)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Optimal KMeans clustering (k=5) for Wisconsin Breast Cancer data')
```

```python
bandwidth = range(1,31)
numclusters = []
results = 0

for b in bandwidth:
    meanshift = MeanShift(bandwidth=b).fit(features_wisconsin.values)
    results = len(np.unique(meanshift.labels_))
    numclusters.append(round(results,2))

plt.plot(bandwidth,numclusters, marker='*',c='blue')
plt.xlabel('bandwidth')
plt.ylabel('Number of clusters')
```

```python
meanshift_optimal = MeanShift(bandwidth=5).fit_predict(features_wisconsin.values)
features_wisconsin["meanshift"] = meanshift_optimal.tolist()
features_wisconsin.head()
```

```python
plt.scatter(features_wisconsin["radius_mean"],features_wisconsin["texture_mean"],c=features_wisconsin.meanshift)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Optimal Mean Shift clustering (bandwidth=5) for Wisconsin Breast Cancer data')
```

```python
from sklearn.cluster import DBSCAN
```

DBSCAN clustering seems to work best for data with dense but irregularly shaped clusters.  By optimizing both epsilon and the minimum number of samples in a cluster, DBSCAN should be flexible enough to correctly cluster both of the datasets.

```python
eps = np.arange(0.5,20,0.5)
numclusters = []
results = 0

for e in eps:
    dbscan = DBSCAN(eps=e,min_samples=300).fit(features_wisconsin.values)
    results = len(np.unique(dbscan.labels_))
    numclusters.append(round(results,2))

plt.plot(eps,numclusters, marker='*',c='blue')
plt.xlabel('epsilon')
plt.ylabel('Number of clusters')
```

```python
dbscan_optimal = DBSCAN(eps=10,min_samples=300).fit_predict(features_wisconsin.values)
features_wisconsin["dbscan"] = dbscan_optimal.tolist()
features_wisconsin.head()
```

```python
plt.scatter(features_wisconsin["radius_mean"],features_wisconsin["texture_mean"],c=features_wisconsin.dbscan)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Optimal DBSCAN clustering (eps=10) for Wisconsin Breast Cancer data')
```

```python
eps = np.arange(0.5,20,0.5)
numclusters = []
results = 0

for e in eps:
    dbscan = DBSCAN(eps=e,min_samples=300).fit(features_insurance.values)
    results = len(np.unique(dbscan.labels_))
    numclusters.append(round(results,2))

plt.plot(eps,numclusters, marker='*',c='blue')
plt.xlabel('epsilon')
plt.ylabel('Number of clusters')
```

```python
dbscan_optimal = DBSCAN(eps=2,min_samples=300).fit_predict(features_insurance.values)
features_insurance["dbscan"] = dbscan_optimal.tolist()
features_insurance.head()
```

```python
plt.scatter(features_insurance["age"],features_insurance["bmi"],c=features_insurance.dbscan)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Optimal DBSCAN clustering (eps=2) for Insurance data')
```

Given that the Breast Cancer dataset should really only have two clusters, none of the clustering techniques performed well.  It is more difficult to assess the performance of the clustering algorithms on the insurance dataset, as the target is a continuous variable.
